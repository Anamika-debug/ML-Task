```python

import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
#data = open("d1.txt","r")
data = pd.read_csv('/content/sample_data/d1.txt', delimiter = "\t")
print(data)
print(data.shape)
plt.scatter(data['Longitude'],data['Latitude'])
plt.xlim(-2.5,0.8)
plt.ylim(-2,1.8)
plt.show()
#x = data.iloc[:,1:3]
#x=np.asarray(x)
#print(x)
kmeans = KMeans(3)
kmeans.fit(data)

label = kmeans.fit_predict(data)
print(label)

# k means
kmeans = KMeans(n_clusters=3, random_state=0)
data['cluster'] = kmeans.fit_predict(data[['Longitude', 'Latitude']])
# get centroids
centroids = kmeans.cluster_centers_
cen_x = [i[0] for i in centroids]
cen_y = [i[1] for i in centroids]
## add to df
data['cen_x'] = data.cluster.map({0:cen_x[0], 1:cen_x[1], 2:cen_x[2]})
data['cen_y'] = data.cluster.map({0:cen_y[0], 1:cen_y[1], 2:cen_y[2]})
# define and map colors
colors = ['Red', 'Green', 'Blue']
data['c'] = data.cluster.map({0:colors[0], 1:colors[1], 2:colors[2]})

import matplotlib.pyplot as plt
plt.scatter(data.Longitude, data.Latitude, c=data.c)
plt.xlim(-2.5,0.8)
plt.ylim(-2,1.8)
plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
data = pd.read_csv('/content/sample_data/5000 Data_points.csv')
print(data)
print(data.shape)
print(data.describe())
plt.scatter(data['Atr1'],data['Atr2'])
plt.xlim(19835,961951)
plt.ylim(51121,970756)
plt.show()
#x = data.iloc[:,1:3]
#x=np.asarray(x)
#print(x)
kmeans = KMeans(3)
kmeans.fit(data)

label = kmeans.fit_predict(data)
print(label)

# k means
kmeans = KMeans(n_clusters=15, random_state=0)
data['cluster'] = kmeans.fit_predict(data[['Atr1', 'Atr2']])
# get centroids
centroids = kmeans.cluster_centers_
cen_x = [i[0] for i in centroids]
cen_y = [i[1] for i in centroids]
## add to df
data['cen_x'] = data.cluster.map({0:cen_x[0], 1:cen_x[1], 2:cen_x[2], 3:cen_x[3],4:cen_x[4],5:cen_x[5],6:cen_x[6],7:cen_x[7],8:cen_x[8],9:cen_x[9],10:cen_x[10],11:cen_x[11],12:cen_x[12],13:cen_x[13],14:cen_x[14]})
data['cen_y'] = data.cluster.map({0:cen_y[0], 1:cen_y[1], 2:cen_y[2],3:cen_y[3],4:cen_y[4],5:cen_y[5],6:cen_y[6],7:cen_y[7],8:cen_y[8],9:cen_y[9],10:cen_y[10],11:cen_y[11],12:cen_y[12],13:cen_y[13],14:cen_y[14]})
# define and map colors
colors = ['Red', 'Green', 'Blue','Cyan','Pink','Orange', 'Yellow','Violet','Magenta','Black','Brown','Gray','Teal','Indigo','Purple']
data['c'] = data.cluster.map({0:colors[0], 1:colors[1], 2:colors[2], 3:colors[3],4:colors[4],5:colors[5],6:colors[6],7:colors[7],8:colors[8],9:colors[9],10:colors[10],11:colors[11],12:colors[12],13:colors[13],14:colors[14]})

import matplotlib.pyplot as plt
plt.scatter(data.Atr1, data.Atr2, c=data.c)
plt.xlim(19835,961951)
plt.ylim(51121,970756)
plt.show()
```
